```python
fhand = open('mbox.txt')
counts = dict()
for line in fhand:
    words = line.split()
    for word in words:
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1
#print(counts)
for line in fhand:
    line = line.rstrip()
    if line.startswith('From:'):
        for lines in fhand:
            lines = lines.rstrip()
            if lines.startswith('To:'):
                print(line + lines)
for key in counts:
    print (key, counts[key])
```
